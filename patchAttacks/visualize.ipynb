```python
result_folder = "./log/apricot/"
gt_folder = result_folder + "gt"
inpainted_folder = result_folder + "inpainted"
```

```python
gt_files = os.listdir(gt_folder)
inpainted_files = os.listdir(inpainted_folder)

f, axarr = plt.subplots(3,2, figsize=(12, 12))
iter = 0
for img_name in gt_files:
    img = cv2.imread(os.path.join(gt_folder, img_name))
    axarr[iter,0].imshow(img)
    axarr[iter,0].axis('off')
    # axarr[0,0].show()

    inpainted = cv2.imread(os.path.join(inpainted_folder, img_name))
    axarr[iter,1].imshow(inpainted)
    axarr[iter,1].axis('off')
    # axarr[0,0].show()
    iter += 1
```
